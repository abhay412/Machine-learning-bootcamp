```python
# creating a white border all around the img
def make_border(image, border_width, color=255):
    width = len(img[0])
    height = len(img)
    for i in range(height):
        if i < border_width or i > height-border_width:
            for j in range(width):
                image[i][j] = color
        else:
            for j in range(border_width):
                image[i][j] = color
                image[i][j + width - border_width] = color
        
    return image
```

```python
img = cv2.imread('Cat_1.jpg', cv2.IMREAD_GRAYSCALE)
img = make_border(img, 20, color=255)
cv2.imwrite("grey_Cat_3.jpg", img)
```

```python
# Rotate the image
img = cv2.imread('Cat_1.jpg')
plt.imshow(img)
```

```python
img.shape
```

```python
num_rows, num_cols = img.shape[:2]
```

```python
rot_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), 30, 0.9)
rot_matrix
```

```python
img_2 = cv2.warpAffine(img, rot_matrix, (num_cols, num_rows))
plt.imshow(img_2)
```
