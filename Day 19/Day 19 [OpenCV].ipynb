```python
# creating a white border all around the img
def createBorder(self, border_width, sides, color=255):
        width = len(img[0])
        height = len(img)
        for i in range(height):
            if (i < border_width and sides['top']) or (i > height-border_width and sides['bottom']):
                for j in range(width):
                    img[i][j] = color
            else:
                for j in range(border_width):
                    if sides['left']:
                        img[i][j] = color
                    if sides['right']:
                        img[i][j + width - border_width] = color
            
        return img
```
