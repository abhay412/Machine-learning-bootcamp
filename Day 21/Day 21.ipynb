### Open CV [Image Manipulation Continued]

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
img = cv2.imread("windows.png", cv2.IMREAD_GRAYSCALE)
```

```python
img.shape
```

```python
row, col = img.shape
```

```python
img_output = np.zeros(img.shape)

for i in range(row):
    for j in range(col):
        offset_x = int(25 * math.cos(2 * 3.142857 * i/180))
        offset_y = 0
        
        if (j+offset_x) < col:
            img_output[i, j] = img[i, j+offset_x]
        else:
            img_output[i, j] = 0

plt.imshow(img_output)
```

```python
img_output = np.zeros(img.shape)

for i in range(row):
    for j in range(col):
        offset_x = int(25 * math.cos(2 * 3.142857 * i/180))
        offset_y = 0
        
        if (j+offset_x) < col:
            img_output[i, j] = img[i, j+offset_x]
        else:
            img_output[i, j] = 0
            
        if (j+offset_x > col//3) and (j+offset_x < col//3 + 10): 
            img_output[i, j] = 255
            

plt.imshow(img_output)
```

```python
img_output = np.zeros(img.shape)

for i in range(row):
    for j in range(col):
        offset_y = int(25 * math.cos(2 * 3.142857 * j/180))
        offset_x = 0
        
        if (i+offset_y) < row:
            img_output[i, j] = img[i+offset_y, j]
        else:
            img_output[i, j] = 0
            
plt.imshow(img_output)            
```

```python
img_output = np.zeros(img.shape)

for i in range(row):
    for j in range(col):
        offset_y = int(25 * math.cos(2 * 3.142857 * j/180))
        offset_x = int(25 * math.cos(2 * 3.142857 * i/180))
        
        if (i+offset_y) < col and (j+offset_x) < row:
            img_output[i, j] = img[i+offset_y, j+offset_x]
        else:
            img_output[i, j] = 0
            
plt.imshow(img_output)
```

## Image Blur

```python
img = cv2.imread("Cat_1.jpg")
plt.imshow(img)
```

```python
rows, cols = img.shape[:2]
```

```python
kernel_idn = np.array([[0,0,0], [0,1,0], [0,0,0]])
kernel_idn
```

```python
kernel_3 = np.ones((3,3))/9
kernel_3
```

```python
kernel_5 = np.ones((5,5))/25
kernel_5
```

```python
kernel_10 = np.ones((10,10))/100
kernel_10
```

```python
img_output = cv2.filter2D(img, -1, kernel_10)
plt.imshow(img_output)
cv2.imwrite("blur_f10.jpg", img_output)
```

```python
kernel_h50 = np.array([[1 for _ in range(50)]]) / 50
kernel_h50
```

```python
img_output = cv2.filter2D(img, -1, kernel_h50)
plt.imshow(img_output)
cv2.imwrite("horizontal_blur.jpg", img_output)
```

```python
kernel_v50 = kernel_h50.reshape((-1, 1))
# kernel_v50
```

```python
img_output = cv2.filter2D(img, -1, kernel_v50)
plt.imshow(img_output)
cv2.imwrite("vertical_blur.jpg", img_output)
```
