### MatplotLib

```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
# Plotng a line using list data
plt.plot([1,2,3,4,5])
plt.ylabel("Y Label")
plt.xlabel("X Label")
```

```python
# Changin plot marks
plt.plot([1,2,3,4,5], 'r+')
# o - For Bullets
# + - For corss
# colors RGB
```

```python
plt.plot([1,2,3,4,5], markersize=15, marker='o', color='blue')
```

```python
# Drawing multiple plots in same img
t = np.arange(0, 5, .2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
```

**Scatter Plot**

```python
# Creating a scattered data
data = {'x': np.arange(50),
        'y': 10 * np.random.randn(50),
        'color': np.random.randint(0, 50, 50),
        'size': np.abs(np.random.randn(50))*100}

plt.scatter('x', 'y', c='color', s='size', data=data)
```

**BAR Charts** and **Sub Plots**

```python
names = ['group-a','group-b','group-c',]
values = [1, 10, 100]

plt.figure(figsize=(9, 3))
plt.suptitle("Sub Plot Example")

plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.subplot(133)
plt.plot(names, values)
```
