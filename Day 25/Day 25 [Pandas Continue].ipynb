```python
#Loading Train.csv
data_frame = pd.read_csv("./Titanic Dataset/titanic/train.csv")
data_frame.head()
```

```python
# Custom Function to apply on our dataframe
def didSurvived(state):
    return "Survived" if state==1 else "Deceased"
```

```python
new_column = data_frame['Survived'].apply(didSurvived)
# new_column
```

```python
# Adding the above new column to our dataframe
data_frame['State'] = new_column
```

**Merging on DataFrames**

```python
d1 = pd.DataFrame({'ASCII':[97, 98, 99, 100, 101], 'Char':['a', 'b', 'c', 'd', 'e']})
d1
```

```python
d2 = pd.DataFrame({'ASCII':[97, 98, 99, 100, 101], 'CAPS':['A', 'B', 'C', 'D', 'E']})
d2
```

```python
# Simillar to a Join function of SQL
pd.merge(d1, d2)
```

```python
d1 = pd.DataFrame({'id':[1,2,3,4,5], 'ASCII':[97,98,99,100,101], 'Char':['a','b','c','d','e']})
d2 = pd.DataFrame({'id':[1,2,3,4,5], 'ASCII':[97,98,99,100,101], 'CAPS':['A','B','C','D','E']})
```

```python
d1
```
